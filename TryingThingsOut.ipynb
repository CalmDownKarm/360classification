```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
from sklearn.cluster import AgglomerativeClustering
from sklearn.neighbors import kneighbors_graph
from sklearn.manifold import TSNE
from fastText import tokenize, load_model
import gensim
df1 = pd.read_csv("./grantnav-20180526072920.csv",encoding='utf-8')
```

Titles aren't that useful, a lot of them are garbled. 

```python
# titles = [title for title in df.Title.unique() if isinstance(title,str)]
import re
# Cleaning Text
descs = [desc.encode('ascii',errors='ignore').decode('utf-8') for desc in df1.Description.unique() if isinstance(desc,str)]
# Strip all hex 
desc_string = ' '.join(descs)
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
desc_tokens = tokenize(desc_string)
desc_tokens = [word.lower() for word in desc_tokens if (word.isalpha() and word.lower() not in stop_words)]
desc_tokens = list(set(desc_tokens))
```

```python
# Word2Vec
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz', binary=True)
vectors = [model[token] for token in desc_tokens if token in model.vocab]
len(vectors)/len(desc_tokens)
# only about 54% of words are in the vocabulary - might not be worth using word2vec
```

```python
# Fast Text Vectors
model = load_model("./wiki.en.bin")
vectors = [model.get_word_vector(word) for word in desc_tokens]
```

```python
clusterer = KMeans(n_clusters=100, n_init=10)
```

```python
clusterer.fit(vectors)
```

```python
df = pd.DataFrame(list(zip(desc_tokens, vectors,clusterer.labels_)),columns = ['word', 'vector','label'])
```

```python
cleaned_descs = []
for desc in descs:
    desc_tokens = tokenize(desc)
    cleaned_descs.append([word.lower() for word in desc_tokens if (word.isalpha() and word.lower() not in stop_words)])
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF    
# At this point I have a list of lists, where each list has the non stop words of that description. 
no_features = 100
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=no_features, stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(descs)
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
nmf = NMF(n_components=15, random_state=1, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf)#15 topics
def get_topics(model, feature_names, no_topwords):
    for topic_id, topic in enumerate(model.components_):
        print('topic %d:' % (topic_id))
        print(' '.join([feature_names[i] for i in topic.argsort()[:-no_topwords-1:-1]]))
get_topics(nmf, tfidf_feature_names, 20)
```

```python
connectivity = kneighbors_graph(vectors, n_neighbors=10, include_self=False)
ward = AgglomerativeClustering(n_clusters=15, connectivity=connectivity,linkage='ward').fit(vectors)
# http://aa.ssdi.di.fct.unl.pt/files/AA-19_notes.pdf
```

```python
df = pd.DataFrame(list(zip(desc_tokens, vectors,ward.labels_)),columns = ['word', 'vector','label'])
```
